```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
print('Features:', iris.feature_names)
print('Target names:',iris.target_names)
```

```python
data = pd.DataFrame(iris.data, columns = iris.feature_names)
data['target'] = iris.target

print(data.head())
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd


california = fetch_california_housing()


data = pd.DataFrame(california.data, columns=california.feature_names)
data['target'] = california.target  # Add the target column


print(data.head())
```

```python
# Plot the relationship between Median Income and Target
sns.scatterplot(x='MedInc', y='target', data=data)
plt.title('Median Income vs. Median House Value')
plt.show()

# Pairplot for selected features
sns.pairplot(data[['MedInc', 'HouseAge', 'AveRooms', 'target']])
plt.show()
```

##### RMSE: Indicates the average error in predicting house prices (in tens of thousands of dollars).
##### R-squared: Shows how well the model explains the variance in the target variable.
